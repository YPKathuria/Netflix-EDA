###**Data Wrangling / Cleaning Process**


---



**Handling Missing Data**



```python
net.columns
```

```python
x=net[['director', 'cast', 'country', 'date_added','rating', 'duration']]      # columns/variables with missing values
x.head(5)
```

```python
net["director"]=net["director"].fillna("Missing_Director")    # NaN of director, cast & country filled
net["cast"]=net["cast"].fillna("Missing_Cast")
net["country"]=net["country"].fillna("Missing_Country")
net.info()
```

```python
net["date_added"]=net["date_added"].fillna("Missing_date")    # NaN of date_added, rating & duration filled
net["rating"]=net["rating"].fillna("Missing_Rating")
net["duration"]=net["duration"].fillna("Missing_Duration")
net.info()
```
